```python
import re
import json
from collections import Counter

class TextProcessor:
    def __init__(self, file_path, output_json='output.json'):
        self.file_path = file_path
        self.output_json = output_json
        self.text = ''
        self.cleaned_text = ''
        self.words = []
        self.word_frequencies = {}

    def read_file(self):
        with open(self.file_path, 'r', encoding='utf-8') as f:
            self.text = f.read()

    def clean_text(self):
        text = re.sub(r'http\S+|www\S+|https\S+', '', self.text)
        text = re.sub(r'\S+@\S+', '', text)
        text = re.sub(r'[^\w\s\u0600-\u06FF]', '', text)
        self.cleaned_text = text

    def remove_stopwords(self):
        # لیست ساده‌ای از کلمات پرتکرار فارسی
        stopwords = set([
            'و', 'در', 'به', 'از', 'که', 'این', 'را', 'با', 'برای', 'آن',
            'یک', 'تا', 'می', 'بر', 'است', 'اما', 'یا', 'هم', 'نیز'
        ])
        words = self.cleaned_text.split()
        self.words = [word for word in words if word not in stopwords]

    def Word_repetition_count(self):
        self.Word_repetition = dict(Counter(self.words))
        with open(self.output_json, 'w', encoding='utf-8') as f:
            json.dump(self.Word_repetition, f, ensure_ascii=False, indent=4)

```
